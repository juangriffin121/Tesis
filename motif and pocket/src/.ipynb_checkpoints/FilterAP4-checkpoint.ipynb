```python
aln_dict = {}
for record in SeqIO.parse('../proccesed/aln-AP4_less_gaps.txt','fasta'): #../proccesed/aln-AP4.txt
    aln_dict[record.id] = record.seq
print(len(aln_dict))
```

```python
# relacionar posiciones en el alineamiento con las posiciones en la seccuencia humana: aln_pos = aln_postition_list[human_pos]
blast_position = 0
aln_postition_list = []
for aln_position,char in enumerate(my_seq_aln):
    if char != '-':
        aln_postition_list.append(aln_position)
        blast_position += 1
```

## Guardar posiciones relevantes

```python
posiciones = [
    [253, 'S'],
    [254, 'F'],
    [255, 'H'],
    [256, 'S'],
    [258, 'V'],
    [260, 'L'],
    [263, 'F'],
    [264, 'E'],
    [279, 'T'],
    [282, 'R']
]
```

```python
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt

# 1) Conservation function
def conservation_at_column(column):
    counts = Counter(column)
    return counts.most_common(1)[0][1] / len(column)

# 2) Compute fullâalignment conservation
seq_ids = list(aln_dict.keys())
n_cols = len(next(iter(aln_dict.values())))
start = 0
end = n_cols
full_cons = [
    conservation_at_column([aln_dict[sid][col] for sid in seq_ids])
    for col in range(start, end) #(n_cols)
]

# 3) Map your human positions and get their conservation
mapped = [aln_postition_list[pos-1] for pos,_ in posiciones]
selected_cons = [full_cons[i - start] for i in mapped]
```

```python
# 4) Plot comparison
plt.figure(figsize=(12,4))
plt.plot(full_cons, label='Alineamiento completo', alpha=0.4)
plt.scatter(np.array(mapped) - start, selected_cons, color='red', label='Posiciones de interÃ©s', s=10)
plt.legend(loc='lower right')

plt.xlabel('PosiciÃ³n en alineamiento')
plt.ylabel('ConservaciÃ³n (identidad)')
plt.title('ConservaciÃ³n: alineamiento completo vs posiciones de interÃ©s')
plt.show()
mean_full = np.mean(full_cons)
mean_sel  = np.mean(selected_cons)
#print(f'ConservaciÃ³n media alineamiento completo: {mean_full*100:.2f}%')
#print(f'ConservaciÃ³n media posiciones de interÃ©s: {mean_sel*100:.2f}%')
```

```python
#se guarda la informacion de los organismos que difieren en alguna posicion de la proteina humana
different_aa_dict = {}
for ID,aln_seq in dict.items(aln_dict):
    different_aa_dict[ID] = []
    for position in posiciones:
        pos = position[0]
        my_aa = position[1]
        seq_aa = aln_seq[aln_postition_list[pos]]
        #print(pos,seq_aa,my_aa,ID)
        if seq_aa != my_aa:
            different_aa_dict[ID].append(pos)
```

```python
# Se calcula el porcentaje de distintos por posicion relevante
pos_dict = {}
for pos in posiciones:
    pos = pos[0]
    pos_dict[pos] = 0
print(pos_dict)
for ID in different_aa_dict:
    lista = different_aa_dict[ID]
    for element in lista:
        pos_dict[element] += 1
print(len(blast_dict))
for pos in pos_dict:
    pos_dict[pos] = 100*pos_dict[pos]/len(blast_dict)
print(pos_dict)
pos_aa_dict = {}
for pos in posiciones:
    pos_aa_dict[pos[0]] =  pos[1]
print(pos_aa_dict)
```

```python
# Se guarda la informacion de los aminoacidos que sustituyen en las posiciones relevantes 
# pos_amount_dict[pos][aa] = numero de organismos en el que la posicion pos es reemplazada por aminoacido aa
pos_amount_dict = {}
for pos in pos_aa_dict:
    pos_amount_dict[pos] = {}
for ID,seq in dict.items(aln_dict):
    lista_pos = different_aa_dict[ID]
    txt = ''
    for pos in lista_pos:
        seq_aa = seq[aln_postition_list[pos]]
        try:
            pos_amount_dict[pos][seq_aa] +=1
        except:
            pos_amount_dict[pos][seq_aa] = 1
print(pos_amount_dict)
```

```python
# calculo de porcentajes ocn la informacion obtenida previamente
pos_percentage_dict = {}
for pos in pos_amount_dict:
    pos_percentage_dict[pos] = {}
    aa_dict = pos_amount_dict[pos]
    total = pos_dict[pos]
    for aa in aa_dict:
        amount = aa_dict[aa]
        percentage = (100*amount/len(blast_dict))
        pos_percentage_dict[pos][aa] = percentage
        print(pos,aa,amount,percentage)
```

```python
complete_both_diff_organism_dict = {}
for App_ID in App_diferent_YKFFE:
    organism = App_org_dict[App_ID[0:-2]]
    motif = str(App_diferent_YKFFE[App_ID])
    AP4_IDS = both_diff_org_dict[organism]
    complete_both_diff_organism_dict[organism] = {'App_IDS' : App_ID ,'motif' : motif}
    for AP4_ID in AP4_IDS:
        different_positions = both_diff_org_dict[organism][AP4_ID]
        try:
            complete_both_diff_organism_dict[organism]['AP4_IDS'][AP4_ID] = different_positions
        except:
            complete_both_diff_organism_dict[organism]['AP4_IDS'] = {AP4_ID : different_positions}
print(complete_both_diff_organism_dict)
```

```python
pos_amount_dict_filt = {}
num_sequences = 0
for organism in complete_both_diff_organism_dict:
    for AP4_ID in complete_both_diff_organism_dict[organism]['AP4_IDS']:
        different_positions = complete_both_diff_organism_dict[organism]['AP4_IDS'][AP4_ID]
        for pos in different_positions:
            seq_aa = different_positions[pos]
            num_sequences += 1
            try:
                pos_amount_dict_filt[pos][seq_aa] += 1
            except:
                try:
                    pos_amount_dict_filt[pos] = {seq_aa : 1}
                except:
                    pos_amount_dict_filt = {pos : {seq_aa : 1}}

print(pos_amount_dict_filt)
```

```python
differences = {}
for App_ID in App_diferent_YKFFE:
    organism = App_org_dict[App_ID[0:-2]]
    print(organism)
    motif = str(App_diferent_YKFFE[App_ID])
    try:
        AP4_IDs = org_dict[organism]
    except:
        for org in org_dict:
            try:
                place = org.find(organism) #busco el nombre en el nombre que puede ser mas largo
                if place != -1:
                    AP4_IDs = org_dict[org]
            except:
                pass
    differences[organism] = {}
    for AP4_ID in AP4_IDs:
        print('\t',AP4_ID)
        AP4_ID = AP4_ID + '.1'
        differences[organism][AP4_ID] = {'App_ID': App_ID , 'motif' : motif , 'positions' : {} }
        seq =  aln_dict[AP4_ID]
        for pos in posiciones:
            true_aa = pos[1]
            pos=pos[0]
            seq_aa = seq[aln_postition_list[pos]]
            differences[organism][AP4_ID]['positions'][pos] = seq_aa
            print('\t\t',motif,pos,'true_aa:',true_aa,seq_aa)
```
