```python
YTSI_span = (my_seq.find('YTSI'),my_seq.find('YTSI')+4)
YENPTY_span = (my_seq.find('YENPTY'),my_seq.find('YENPTY')+6)
YKFFE_span = (my_seq.find('YKFF'),my_seq.find('YKFF')+5)
print(YKFFE_span)
```

```python
different_motifs = []
for ID,seq in dict.items(blast_dict):
    motif = seq[YTSI_span[0]:YTSI_span[1]]
    if motif != 'YTSI':
        gap = False
        for aa in motif:
            if aa == '-':
                gap = True
        if not gap:
            different_motifs.append(ID)
            print(f'{motif}:{ID}')
print(different_motifs)
```

```python
for ID,seq in dict.items(blast_dict):
    motif = seq[YENPTY_span[0]:YENPTY_span[1]]
    if motif != 'YENPTY':
        gap = False
        for aa in motif:
            if aa == '-':
                gap = True
        if not gap:
            different_motifs.append(ID)
            print(f'{motif}:{ID}')
```

```python
different_motifs_YKFFE = {}
for ID,seq in dict.items(blast_dict):
    motif = seq[YKFFE_span[0]:YKFFE_span[1]]
    if motif != 'YKFFE':
        gap = False
        for aa,motif_aa in zip(motif,'YKFFE'):
            if aa == '-' and motif_aa != '-':
                gap = True
        if not gap:
            different_motifs.append(ID)
            different_motifs_YKFFE[ID] = str(motif)
            print(f'{motif}:{ID}')
```

```python
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt

# 1) Conservation function
def conservation_at_column(column):
    counts = Counter(column)
    return counts.most_common(1)[0][1] / len(column)

# 2) Compute full‐alignment conservation
seq_ids = list(blast_dict.keys())
n_cols = len(next(iter(blast_dict.values())))
start = 0
end = n_cols
full_cons = [
    conservation_at_column([blast_dict[sid][col] for sid in seq_ids])
    for col in range(start, end) #(n_cols)
]

# 3) Map your human positions and get their conservation
mapped = [i for i in range(1699,1705 + 1)]
selected_cons = [full_cons[i - start] for i in mapped]

# 4) Plot comparison
plt.figure(figsize=(12,4))
plt.plot(full_cons, label='Alineamiento completo', alpha=0.5)
plt.scatter(np.array(mapped) - start, selected_cons, color='red', label='Posiciones de interés')

plt.legend()
plt.xlabel('Posición en alineamiento')
plt.ylabel('Conservación (identidad)')
plt.title('Conservación: alineamiento completo vs posiciones de interés')
plt.show()
mean_full = np.mean(full_cons)
mean_sel  = np.mean(selected_cons)

print(f'Conservación media alineamiento completo: {mean_full*100:.2f}%')
print(f'Conservación media posiciones de interés: {mean_sel*100:.2f}%')
```
