```python
import pandas
import json
from Bio import SeqIO
from Bio import GenBank
```

```python
blast_dict = {}
for record in SeqIO.parse('../proccesed/aln-App_corrected.txt','fasta'):
    blast_dict[record.id] = record.seq
print(len(blast_dict))
```

```python
YTSI_span = (my_seq.find('YTSI'),my_seq.find('YTSI')+4)
YENPTY_span = (my_seq.find('YENPTY'),my_seq.find('YENPTY')+6)
YKFFE_span = (my_seq.find('YKFF'),my_seq.find('YKFF')+5)
print(YKFFE_span)
```

```python
different_motifs = []
for ID,seq in dict.items(blast_dict):
    motif = seq[YTSI_span[0]:YTSI_span[1]]
    if motif != 'YTSI':
        gap = False
        for aa in motif:
            if aa == '-':
                gap = True
        if not gap:
            different_motifs.append(ID)
            print(f'{motif}:{ID}')
print(different_motifs)
```

```python
for ID,seq in dict.items(blast_dict):
    motif = seq[YENPTY_span[0]:YENPTY_span[1]]
    if motif != 'YENPTY':
        gap = False
        for aa in motif:
            if aa == '-':
                gap = True
        if not gap:
            different_motifs.append(ID)
            print(f'{motif}:{ID}')
```

```python
different_motifs_YKFFE = {}
for ID,seq in dict.items(blast_dict):
    motif = seq[YKFFE_span[0]:YKFFE_span[1]]
    if motif != 'YKFFE':
        gap = False
        for aa,motif_aa in zip(motif,'YKFFE'):
            if aa == '-' and motif_aa != '-':
                gap = True
        if not gap:
            different_motifs.append(ID)
            different_motifs_YKFFE[ID] = str(motif)
            print(f'{motif}:{ID}')
```

```python
print(different_motifs)
```

```python
for i,element1 in enumerate(different_motifs):
    for j,element2 in enumerate(different_motifs):
        if i != j:
            if element1 == element2:
                print(i,j,'duplicado')
```

```python
filt_dict = {}
txt = ''
for ID in different_motifs:
    seq = blast_dict[ID]
    txt = txt + f'>{ID}\n'
    txt = txt + f'{seq}\n'
    print(seq[YTSI_span[0]:YTSI_span[1]],'~~~',seq[YENPTY_span[0]:YENPTY_span[1]],'~~~',seq[YKFFE_span[0]:YKFFE_span[1]])
with open('../proccesed/App_filtered.txt','w') as f:
    f.write(txt)
```
