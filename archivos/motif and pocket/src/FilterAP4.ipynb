```python
def graph_differences(dict_organisms,AP4_aln_dict): # dict_organisms 
    different_aa_dict = {}
    for ID,aln_seq in dict.items(AP4_aln_dict):
        different_aa_dict[ID] = []
        for position in posiciones:
            pos = position[0]
            my_aa = position[1]
            seq_aa = aln_seq[aln_postition_list[pos]]
            if seq_aa != my_aa:
                different_aa_dict[ID].append(pos)

    pos_dict = {}
    for pos in posiciones:
        pos = pos[0]
        pos_dict[pos] = 0
    for ID in different_aa_dict:
        lista = different_aa_dict[ID]
        for element in lista:
            pos_dict[element] += 1
    for pos in pos_dict:
        pos_dict[pos] = 100*pos_dict[pos]/len(blast_dict)
    pos_aa_dict = {}
    for pos in posiciones:
        pos_aa_dict[pos[0]] =  pos[1]
    
    pos_amount_dict = {}
    for pos in pos_aa_dict:
        pos_amount_dict[pos] = {}
    for ID,seq in dict.items(AP4_aln_dict):
        lista_pos = different_aa_dict[ID]
        txt = ''
        for pos in lista_pos:
            seq_aa = seq[aln_postition_list[pos]]
            try:
                pos_amount_dict[pos][seq_aa] +=1
            except:
                pos_amount_dict[pos][seq_aa] = 1
    
    pos_amount_dict_filt = {}
    for organism in dict_organisms:
        AP4_IDS = dict_organisms[organism]['AP4_IDS']
        for AP4_ID in AP4_IDS:
            different_positions = dict_organisms[organism]['AP4_IDS'][AP4_ID]
            for pos in different_positions:
                seq_aa = different_positions[pos]
                try:
                    pos_amount_dict_filt[pos][seq_aa] += 1
                except:
                    try:
                        pos_amount_dict_filt[pos] = {seq_aa : 1}
                    except:
                        pos_amount_dict_filt = {pos : {seq_aa : 1}}

    pos_percentage_dict = {}
    for pos in pos_amount_dict:
        pos_percentage_dict[pos] = {}
        aa_dict = pos_amount_dict[pos]
        total = pos_dict[pos]
        for aa in aa_dict:
            amount = aa_dict[aa]
            percentage = (100*amount/len(blast_dict))
            pos_percentage_dict[pos][aa] = percentage
    
    fig,ax = plt.subplots(figsize=(15,5))
    #ax1
    print('AAs diferentes en DataSet sin filtrar')
    ax.plot([0,100],alpha = 0)
    x=0
    for pos in pos_dict:
        aa = pos_aa_dict[pos]
        percentage = pos_dict[pos]
        y = 0
        for diff_aa in pos_percentage_dict[pos]:
            aa_percentage = pos_percentage_dict[pos][diff_aa]
            ax.add_patch(Rectangle((x+0.1,1*y),0.9,aa_percentage,alpha = 1,facecolor=color_dict[diff_aa]))
            ax.text(x+0.4,y+aa_percentage/2,diff_aa)
            y += aa_percentage
        ax.text(x+0.4,-5,f'{aa}{pos+1}')
        #ax.add_patch(Rectangle((x+0.1,0),0.9,percentage,alpha = 0.5,facecolor=(0.,0.,1.,1.)))
        x += 1
    x = 0
    for pos in pos_aa_dict:
        aa = pos_aa_dict[pos]
        percentage = pos_dict[pos]
        ax.add_patch(Rectangle((x+0.3,percentage),0.5,100-percentage,alpha = 1,facecolor=(0.7,0.7,0.7)))
        ax.text(x+0.4,percentage+(100-percentage)/2,aa)
        x += 1

    plt.show()
    #ax2
    #los porcentajes de aminoacidos distintos en los que tienen YKFFE distinto
    print('AAs diferentes en DataSet filtrado por condicion')
    fig,ax = plt.subplots(figsize=(15,5))
    ax.plot([0,10],alpha = 0)
    x=0
    for pos in pos_aa_dict:
        aa = pos_aa_dict[pos]
        y = 0
        try:
            for diff_aa in pos_amount_dict_filt[pos]:
                aa_percentage = pos_amount_dict_filt[pos][diff_aa]
                ax.add_patch(Rectangle((x+0.1,1*y),0.9,aa_percentage,alpha = 1,facecolor=color_dict[diff_aa]))
                ax.text(x+0.4,y+aa_percentage/2,diff_aa)
                y += aa_percentage
        except:
            pass
        ax.text(x+0.4,-1,f'{aa}{pos+1}')
        x +=1
    plt.show()
```

```python
def filter_org_dict_App(org_dict,aln_dict):
    filtered_dict = {}
    for org in org_dict:
        IDS = org_dict[org]['App_IDS']
        print(org,IDS)
        for ID in aln_dict:
            if ID in IDS:
                filtered_dict[org]['AP4_IDS'] = org_dict[org]['AP4_IDS']
                try:
                    filtered_dict[org]['App_IDS'].append(ID)
                except:
                    filtered_dict[org]['App_IDS'] = [ID]
    return filtered_dict

def filter_org_dict_AP4(org_dict,aln_dict):
    filtered_dict = {}
    for org in org_dict:
        IDS = org_dict[org]['AP4_IDS']
        for ID in aln_dict:
            if ID in IDS:
                filtered_dict[org]= {'App_IDS' : org_dict[org]['App_IDS']}
                try:
                    filtered_dict[org]['AP4_IDS'][ID] = org_dict[org]['AP4_IDS'][ID]
                except:
                    filtered_dict[org]['AP4_IDS'] = {ID : org_dict[org]['AP4_IDS'][ID]}
    return filtered_dict
i = 0
for element in AP4_diff_App_equal_org_dict:
    if i == 1:
        print(element,':')
        for entry in AP4_diff_App_equal_org_dict[element]:
            print('\t',entry,':')
            print('\t','\t',AP4_diff_App_equal_org_dict[element][entry])
    i += 1
print()

i = 0
for element in filter_org_dict_AP4(AP4_diff_App_equal_org_dict,aln_dict_AP4_mammals):
    if i == 1:
        print(element,':')
        for entry in AP4_diff_App_equal_org_dict[element]:
            print('\t',entry,':')
            print('\t','\t',AP4_diff_App_equal_org_dict[element][entry])
    i += 1
    
```
